# Car Type Count Calculation

```python
def cartype (value):
    if value <= 15:
        return 'low'
    elif 15 < value <= 25:
        return 'medium'
    else:
        return 'high'
```

```python
df['cartype'] = df['car'].apply(cartype)
```

```python
df.head()
```

```python
df['cartype'].value_counts()
```

# Bus Count Index Retrieval
